```python
# Import necessary libraries
import pandas as pd
from pycaret.anomaly import *
```

```python
# Loading the dataset
data = pd.read_csv('BankChurners.csv')
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
# Setup PyCaret for Anomaly Detection
exp_ano = setup(data, normalize=True, session_id=123)
```

```python
# Create an anomaly detection model
model = create_model('iforest')
```

```python
# Evaluate the best model (plot anomalies)
plot_model(model)
```

```python
# Predict anomalies on the dataset
predictions = predict_model(model, data=data)
```
